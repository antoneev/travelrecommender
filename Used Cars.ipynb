### Removing unwanted columns 

```python
df2 = df1.drop(['location_region','feature_0','feature_1','feature_2','feature_3','feature_4','feature_5','feature_6','feature_7','feature_8','feature_9'],axis='columns')
```

```python
df2
```

### Removing NAs

```python
df3 = df2.dropna()
df3.isnull().sum()
```

### Removing observations not in english

#### By removing the non English manufacturer names this will handle removing any non English model name which aren't a common model name assoicated with English cars

```python
len(df3.manufacturer_name.unique())
```

```python
df3 = df3[df3['manufacturer_name'] != 'ВАЗ']
df3 = df3[df3['manufacturer_name'] != 'ГАЗ']
df3 = df3[df3['manufacturer_name'] != 'ЗАЗ']
df3 = df3[df3['manufacturer_name'] != 'Москвич']
df3 = df3[df3['manufacturer_name'] != 'УАЗ']
```
