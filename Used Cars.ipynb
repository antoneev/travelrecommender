# ANOVA TEST and TWO SAMPLE T-TEST

```python
df6.price_usd.groupby(df6.engine_fuel).mean()
```

```python
from scipy.stats import f_oneway #ANOVA TEST
```

```python
from scipy.stats import ttest_ind #T-Test
```

```python
diesel = df6[df6.engine_fuel=='diesel']    #ANOVA TEST FOR ENGINE_FUEL
gas = df6[df6.engine_fuel=='gas']
gasoline = df6[df6.engine_fuel=='gasoline']
hybrid_diesel = df6[df6.engine_fuel=='hybrid-diesel']
hybrid_petrol = df6[df6.engine_fuel=='hybrid-petrol']
```

```python
f_oneway(diesel.price_usd,gas.price_usd,gasoline.price_usd,hybrid_diesel.price_usd,hybrid_petrol.price_usd) 
```

```python
df6.price_usd.groupby(df6.color).mean()
```

```python
silver = df6[df6.color == 'silver']
blue = df6[df6.color == 'blue']
red = df6[df6.color == 'red']
black = df6[df6.color == 'black']
grey = df6[df6.color == 'grey']
other = df6[df6.color == 'other']
brown = df6[df6.color == 'brown']
white = df6[df6.color == 'white']
green = df6[df6.color == 'green']
violet = df6[df6.color == 'violet']
orange = df6[df6.color == 'orange']
yellow = df6[df6.color == 'yellow']
```

```python
f_oneway(silver.price_usd,blue.price_usd,red.price_usd,black.price_usd,grey.price_usd,other.price_usd,brown.price_usd,white.price_usd,green.price_usd,violet.price_usd,orange.price_usd,yellow.price_usd)
#ANOVA TEST FOR COLOR
```

```python
df6.price_usd.groupby(df6.transmission).mean()
```

```python
automatic = df6[df6.transmission == 'automatic']
mechanical = df6[df6.transmission == 'mechanical']
```

```python
f_oneway(automatic.price_usd,mechanical.price_usd) #ANOVA TEST FOR TRANSMISSION
```

```python
ttest_ind(automatic.price_usd,mechanical.price_usd) #T-TEST FOR TRANSMISSION
```
